<div class="alert alert-block alert-warning">
    <b>Exercise 1:</b> Write a function that computes the level of output generated by a constant returns to scale Cobb-Douglas production function, i.e., such that it computes $A\cdot K^\alpha \cdot L^{1-\alpha}$.
</div>

```python
def CobbDouglas(K=1, L=1, A=1, alpha=0.3):
    '''
    This function computes the level of output of a Cobb-Douglas production function.
    Given values of K, L, A, and alpha it returns $A\cdot K^\alpha \cdot L^{1-\alpha}$.
    Parameters:
    -----------
    K: float, integer or any other number, default=1
    L: float, integer or any other number, default=1
    A: float, integer or any other number, default=1
    alpha: float between (0,1), default=0.3
    
    
    Returns:
    --------
    out = A * K**alpha * L**(1-alpha)
    
    Example:
    >>> out = CobbDouglas(4, 9, 2, 1/2)
    >>> out
        12
    '''
    return A * K**alpha * L**(1-alpha) 
```

```python
CobbDouglas(4, 9, 2, 1/2)
```

    <b>Exercise 2:</b> Use the function you created and plot the production function as a function of $K$ for given values of $A$ and $L$. <b>Hint:</b> Use the <code>np.linspace(0, 5, 100)</code> function to create an array of values of $K$ with 100 point between 0 and 5.
</div>

```python
#creating a data set for K, to graph the line
k = np.linspace(0, 5, 100)
k
```

```python
#computes the matching Cobb Douglas value for the k set
CobbDouglas(K=k, L=9, A=2, alpha=1/2)
```

```python
#graphs the line where K is changing
fig, ax = plt.subplots(figsize=(8,5))
x = k 
y1 = CobbDouglas(K=k, L=9, A=2, alpha=1/2)

f1 = plt.plot(x,y1, c='r', marker='o', linestyle=':', label=r'$y=x^2$')

ax.set_xlabel(r'k = capital per worker', fontsize=18)
ax.set_ylabel(r'Y = GDP', fontsize=18)
ax.set_title(r'Basic Cobb Douglas', fontsize=24)

plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD.png')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD.pdf')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD.jpg')
```

<div class="alert alert-block alert-warning">
    <b>Exercise 3:</b> Show with a plot the effect of increasing $A$ from 1 to 2, 4, or 5.
</div>

```python
fig, ax = plt.subplots(figsize=(8,5))
x = k 
y2 = CobbDouglas(K=k, L=9, A=1, alpha=1/2)
z = CobbDouglas(K=k, L=9, A=2, alpha=1/2)
g = CobbDouglas(K=k, L=9, A=4, alpha=1/2)
h = CobbDouglas(K=k, L=9, A=5, alpha=1/2)

f2 = plt.plot(x,y2, c='r', marker='o', linestyle=':', label=r'$A=1$')
f2 = plt.plot(x,z, c='blue', marker='o', linestyle=':', label=r'$A=2$')
f2 = plt.plot(x,g, c='g', marker='o', linestyle=':', label=r'$A=4$')
f2 = plt.plot(x,h, c='y', marker='o', linestyle=':', label=r'$A=5$')

ax.set_xlabel(r'k = capital per worker', fontsize=18)
ax.set_ylabel(r'Y = GDP', fontsize=18)
ax.set_title(r'Cobb Douglas with Shifting Technology', fontsize=24)
ax.legend(fontsize=10)

plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-A.png')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-A.pdf')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-A.jpg')
```

<div class="alert alert-block alert-warning">

```python
fig, ax = plt.subplots(figsize=(8,5))
x = k 
l = CobbDouglas(K=k, L=1, A=1, alpha=1/2)
m = CobbDouglas(K=k, L=2, A=1, alpha=1/2)
n = CobbDouglas(K=k, L=4, A=1, alpha=1/2)
o = CobbDouglas(K=k, L=5, A=1, alpha=1/2)

f3 = plt.plot(x,l, c='r', marker='o', linestyle=':', label=r'$A=1$')
f3 = plt.plot(x,m, c='b', marker='o', linestyle=':', label=r'$A=2$')
f3 = plt.plot(x,n, c='g', marker='o', linestyle=':', label=r'$A=4$')
f3 = plt.plot(x,o, c='y', marker='o', linestyle=':', label=r'$A=5$')

ax.set_xlabel(r'k = capital per worker', fontsize=18)
ax.set_ylabel(r'Y = GDP', fontsize=18)
ax.set_title(r'Cobb Douglas with Shifting Labor', fontsize=24)
ax.legend(fontsize=10)

plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-L.png')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-L.pdf')
plt.savefig('C:\\Users\\skand\\Desktop\\Lexi\'s Files\\College\\College Sem 7\\ECO4362-EcoGrowth\\notebooks\\Solutions-Function-Package/BCD-L.jpg')
```
